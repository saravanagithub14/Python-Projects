```python

#####  Defining the functions to take player names as input.
def main():
    print ('Welcome to the game....')
    player_one =input ("Enter player-1 name ")
    player_two =input ("Enter player-2 name ")
    print(f"Thank you for joining {player_one} and {player_two}")
main()
#  The board list will act as the template for the game board in "print_board" function.

board= ["-","-","-",
       "-","-","-",
       "-","-","-"]
#  Global Variables were initialized for future purpose.
current_player = "X"
winner = None
game_running = True 

#  Game board with index position for the gameplay.

def print_board(board):
    print(board[0] + " | " + board[1] + " | "+ board[2])
    print(board[3] + " | " + board[4] + " | "+ board[5])
    print(board[6] + " | " + board[7] + " | "+ board[8])

#  Function definedd to take player input 

def player_input(board):
    inp=int(input("input any number between 1 and 9 -" ))
    if inp>=1 and inp<=9 and board[inp-1] == "-" :
        board[inp-1]=current_player
    else:
        print("no place to play ")
        
#  There are 3 winning conditions in the traditional Tic-Tac-Toe - Vertical, Horizontal and Diagonal winners.
#  Defining horizontal winner. 
def horizontal_winner(board):
    global winner
    if board[0]==board[1]==board[2]and board[1]!= "-" :
        winner=board[0]
        return True
    elif board[3]==board[4]==board[5]and board[3]!="-":
        winner=board[0]
        return True 
        
    elif board[6]==board[7]==board[8]and board[7]!="-":
        winner=board[0]
        return True
        
#  Defining vertical winner. 
def vertical_winner(board):
    global winner
    if board[0]==board[3]==board[6]and board[3]!="-":
        winner=board[0]
        return True
    elif board[1]==board[4]==board[7]and board[4]!="-":
        winner=board[0]
        return True 
        
    elif board[2]==board[5]==board[8]and board[5]!="-":
        winner=board[0]
        return True
       
# Definng Diagonal winner. 
def diagonal_winner(board):
    global winner
    if board[0]==board[4]==board[8]and board[4]!="-":
        winner=board[0]
        return True
    elif board[2]==board[4]==board[6]and board[6]!="-":
        winner=board[0]
        return True 
#  Defining conditions for tie, where both the playres are winners.
def tie(board):
    global game_running
if "-" not in board:
    print_board(board)
    print("its a tie")
    game_running=false
    
#  Defining condition to declare the winner    
def check_winner(board):
    if horizontal_winner(board) or vertical_winner(board) or diagonal_winner(board):
        print(f"The winner is.........{winner}") 

#  Defining a function to switch the players after every turn. 
def switch_player():
    global current_player
    if current_player=="X":
        current_player="O"
    else:
        current_player="X"
        game_running=False  
        
#  Defining a loop to keep the game on till someone wins or the board get filled  
while game_running:
    print_board(board)
#  if any input comes for winner break the programm    
    if winner != None:
        break
    player_input(board)
    check_winner(board)
    tie(board)
    switch_player()
    

    

```
