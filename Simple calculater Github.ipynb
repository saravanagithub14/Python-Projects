```python
from tkinter import *
from tkinter.ttk import *

def click(values):
    current = display.get()
    display.delete(0,END)
    display.insert(0,str(current)+str(values))
    
def evaluate():
    try:
        a = display.get()
        result = eval(a)
        display.delete(0,END)
        display.insert(0,str(result))
    except:
        display.delete(0,END)
        display.insert(0,"ERROR") 
        
def clear():
    display.delete(0,END)
    

window = Tk()
window.title ("Simple Calculator")
window.geometry ('300x465')
display = Entry(window,width=400,font = ('Arial',25))
display.place(x=0, y=0, width=300, height=90)
btn_a = Button(window, text = "(",command = lambda : click('('))
btn_a.place(x=0,y=90,width=75,height=75)
btn_b = Button(window, text = ")",command = lambda : click(')'))
btn_b.place(x=75,y=90,width=75,height=75)
btn_c = Button(window, text = "%",command = lambda : click('%'))
btn_c.place(x=150,y=90,width=75,height=75)
btn_d = Button(window, text = "AC",command = clear)
btn_d.place(x=225,y=90,width=75,height=75)
btn_7 = Button(window, text = "7",command = lambda : click(7))
btn_7.place(x=0,y=165,width=75,height=75)
btn_8 = Button(window, text = "8",command = lambda : click(8))
btn_8.place(x=75,y=165,width=75,height=75)
btn_9 = Button(window, text = "9",command = lambda : click(9))
btn_9.place(x=150,y=165,width=75,height=75)
btn_e = Button(window, text = "รท",command = lambda : click('/'))
btn_e.place(x=225,y=165,width=75,height=75)
btn_4 = Button(window, text = "4",command = lambda : click(4))
btn_4.place(x=0,y=240,width=75,height=75)
btn_5 = Button(window, text = "5",command = lambda : click(5))
btn_5.place(x=75,y=240,width=75,height=75)
btn_6 = Button(window, text = "6",command = lambda : click(6))
btn_6.place(x=150,y=240,width=75,height=75)
btn_x = Button(window, text = "x",command = lambda : click('*'))
btn_x.place(x=225,y=240,width=75,height=75)
btn_1 = Button(window, text = "1",command = lambda : click(1))
btn_1.place(x=0,y=315,width=75,height=75)
btn_2 = Button(window, text = "2",command = lambda : click(2))
btn_2.place(x=75,y=315,width=75,height=75)
btn_3 = Button(window, text = "3",command = lambda : click(3))
btn_3.place(x=150,y=315,width=75,height=75)
btn_f = Button(window, text = "-",command = lambda : click('-'))
btn_f.place(x=225,y=315,width=75,height=75)
btn_0 = Button(window, text = "0",command = lambda : click(0))
btn_0.place(x=0,y=390,width=75,height=75)
btn_dot = Button(window, text = ".",command = lambda : click('.'))
btn_dot.place(x=75,y=390,width=75,height=75)
btn_g = Button(window, text = "=",command = evaluate)
btn_g.place(x=150,y=390,width=75,height=75)
btn_g = Button(window, text = "+",command = lambda : click('+'))
btn_g.place(x=225,y=390,width=75,height=75)
window.mainloop()
```
